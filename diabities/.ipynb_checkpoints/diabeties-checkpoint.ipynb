```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dh = pd.read_csv("C:/Users/aryan/OneDrive/Desktop/health_app/diabities/diabetes.csv")
```

```python
dh.head()
```

```python
dh.info()
```

```python
dh["Outcome"].value_counts()
```

```python
x = dh.drop(columns="Outcome",axis = 1)
```

```python
y = dh["Outcome"]
```

```python
y.head()
```

```python
x.head()
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(x)
```

```python
standardx = scalar.transform(x)
```

```python
standardx
```

```python
x = standardx
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=430,stratify=y)
```

```python
clasifirer = svm.SVC(kernel="linear")
```

```python
clasifirer.fit(x_train,y_train)
```

```python
accuacyScore_x_train =clasifirer.predict(x_train) 
```

```python
trainingaccuracy = accuracy_score(accuacyScore_x_train,y_train)
trainingaccuracy
```

```python
accuracyScoe_x_test = clasifirer.predict(x_test)
traningaccuracemain = accuracy_score(accuracyScoe_x_test,y_test)
traningaccuracemain
```
