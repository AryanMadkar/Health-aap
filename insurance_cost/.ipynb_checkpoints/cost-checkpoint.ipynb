```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dh = pd.read_csv("C:/Users/aryan/OneDrive/Desktop/health_app/insurance_cost/insurance.csv")
```

```python
dh.head()
```

```python
x = dh.drop(columns="charges",axis=1)
y = dh["charges"]
```

```python
x.head()
```

```python
dh["region"].value_counts()
```

```python
dh["sex"].value_counts()
```

```python
dh["sex"] = dh["sex"].replace({"male":1,"female":0})
```

```python
dh["region"] = dh["region"].replace(
    {"southeast": 0, "southwest": 1, "northwest": 2, "northeast": 3}
)
dh["smoker"] = dh["smoker"].replace({"yes": 1, "no": 0})
```

```python
dh.head()
```

```python
x = dh.drop(columns="charges",axis=1)
y=dh["charges"]
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
scalar.fit(x)
```

```python
x = scalar.transform(x)
```

```python
randomstate = None
best_score = 0
from sklearn.metrics import accuracy_score
```

```python
for randomstate in range(1,1001):
    x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y,test_size=0.2,random_state=randomstate)
    model = LinearRegression()
    model.fit(x_train,y_train)
    x_train_prediction = model.predict(x_train)
    tranningacc = accuracy_score(x_train_prediction,y_train)
    if tranningacc>best_score:
        best_score = tranningacc
        best_randomstate = randomstate
print(f"best random state:{best_randomstate}")
print(f"best score:{best_score}")
```
